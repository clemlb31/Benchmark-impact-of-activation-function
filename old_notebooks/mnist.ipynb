```python
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
```

```python
# mnist download
(mnist_X_train_full, mnist_y_train_full), (mnist_X_test, mnist_y_test) = (keras.datasets.mnist.load_data())
```

```python
X_train_scaled,X_val_scaled,X_test_scaled,y_train,y_val,y_test = mnist_preprocessing(mnist_X_train_full, mnist_y_train_full, mnist_X_test, mnist_y_test,test_size=0.2,val_size=0.2)
```

```python
model = mnist_model(mode='relu').to(device)
criterion = nn.CrossEntropyLoss()
learning_rate = 0.001
num_epochs = 30
batch_size = 64


optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

train_dataset = TensorDataset(torch.tensor(X_train_scaled, dtype=torch.float32).unsqueeze(1).to(device), torch.tensor(y_train, dtype=torch.long).view(-1).to(device))
val_dataset = TensorDataset(torch.tensor(X_val_scaled, dtype=torch.float32).unsqueeze(1).to(device), torch.tensor(y_val, dtype=torch.long).view(-1).to(device))
test_dataset = TensorDataset(torch.tensor(X_test_scaled, dtype=torch.float32).unsqueeze(1).to(device), torch.tensor(y_test, dtype=torch.long).view(-1).to(device))

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
history = train_model(model, criterion, optimizer, num_epochs,train_loader, val_loader, device)
```
